```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(filename,sep):
  data = pd.read_csv(filename,sep=sep)
  return data
```

```python
path = "../data/j120.sm.rs/j1201_1.sm.csv"
data = load_data(path,sep=" ")
keys = data.keys()
keys
```

### get data

```python
job_no = data[keys[0]]
begin  = data[keys[1]]
end    = data[keys[2]]
```

```python
fig,aix = plt.subplots()
for index  in range(len(job_no)):
    current_job_no = job_no[index]
    current_begin  = begin[index]
    current_end    = end[index]
    aix.plot([begin,end],[job_no,job_no])
fig.show()
```

```python
path = "../data/j30.sm.rs/j301_1.sm.csv"
data = load_data(path," ")
keys = data.keys()
```

```python
job_no = data[keys[0]]
begin  = data[keys[1]]
end    = data[keys[2]]
```

```python
fig,aix = plt.subplots()
for index in range(len(job_no)):
    aix.plot([begin[index],end[index]],[job_no[index]]*2)
    aix.annotate(str(begin[index]),(begin[index],job_no[index]))
    aix.annotate(str(end[index]),(end[index],job_no[index]))
    # aix.annotate("("+str(job_no[index])+")",((begin[index]+end[index])/2,job_no[index]))
fig.savefig("../data/j30.sm.rs/j301_1.png",dpi=300)
```

```python
path = "../data/j30.sm.rs/j301_12.sm.csv"
data = load_data(path,sep = " ")
keys = data.keys()
```
